```python
class ClearScreen:
    
    def __init__(self):
    
        self.clear = os.system('cls' if os.name=='nt' else 'clear')
```

```python
def countFiles(path, extension="*.txt"): 
    
    # Construir el patrón de búsqueda
    patron = os.path.join(path, extension)
    
    # Obtener la lista de archivos que coinciden con el patrón
    archivos = glob.glob(patron)
    
    return len(archivos)
```

```python
if __name__ == "__main__":
    
    path = os.path.join(os.getcwd(), "corpus")
    
     # Extensión de archivo que deseas contar
    extension = "*.txt"
    
    # Imprimir el número de archivos encontrados
    print(f"El número de archivos *.txt en el directorio {path} es: {countFiles(path, extension)}")
    
    band = False
    while not band:
        # Define the size of the list    
        try:
            iteracion = int(input(f"Elija un numero entre 0 y {countFiles(path, extension) - 1}: "))
            if iteracion > 0 and iteracion <= countFiles(path, extension):
                print(f"Se recorrerá hasta la iteración {iteracion}")
                band = True
            else:
                print(f"Usted debe ingresar un valor entre 0 y {countFiles(path, extension) - 1}\n")
                input("Presione alguna tecla para seguir...")
                ClearScreen().clear
                
        except Exception as e:
            print(str(e)+"\n")
            print("Usted debe ingresar un valor númerico \n")
            input("Presione alguna tecla para seguir...")
            ClearScreen().clear
            
            
    datos = []
    
    for i, value in enumerate(glob.glob(os.path.join(path, extension))):      
              
        print("File: " + str(i))
        
        file = open(value, encoding="utf-8-sig")
        
        text = file.read()
        
        file.close()
        
        filename = os.path.basename(value)
        
        datos.append([i, filename, text])
        
        
         
        if i == iteracion - 1:
            
            break 
    
    df = pd.DataFrame(datos, columns=["index", "filename", "text"])
    
    if os.path.exists("intrinsic.xlsx"):

        os.remove("intrinsic.xlsx")
    
    df.to_excel("intrinsic.xlsx", index=False)
```
